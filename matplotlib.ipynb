# MatplotLib Tutorial
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
# compute the x and y coordinates for points on a sine curve

x = np.arange(0, 4 * np.pi,0.1)
y = np.sin(x)
plt.title('sin wave form')

# Plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
# subplot()
# compute the x and y coordinates for points on sine and cosine curves

np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# setting up a subplot that havs height 2 and width 1,
# and set the first such subplot as active
plt.subplot(2,1,1)

# make the first subplot
plt.plot(x,y_sin)
plt.title('Sin')

# make the second subplot and set it as active
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')

# show the figure
plt.show()
```

```python
# Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,align = 'center')
plt.bar(x2,y2, color = 'g')
plt.title('Bar Graph')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')

plt.show()
```

# Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=20)
# by default bins is equal to 10
plt.title("Histogram")
plt.show()
```

# Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
# patch_artist is for the color
```

# Pie Chart

```python
# Data to plot
labels = 'Python', 'C++','Ruby','Java'
sizes = [125,130,245,210]
colors = ['Gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0) # explode first slice

# plot

plt.pie(sizes, explode=explode, labels=labels,colors=colors, autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```

```python
# Data to plot
labels = 'Python', 'C++','Ruby','Java'
sizes = [125,130,245,210]
colors = ['Gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0.4,0.3,0.2) # explode first slice

# plot

plt.pie(sizes, explode=explode, labels=labels,colors=colors, autopct='%1.f%%',shadow=False)
# autopct determines the format, and without shadow it will be plain
plt.axis('equal')
plt.show()
```
